## IMPLEMENTACIÓN DEL GENERADOR PARA ESCALAS MENORES

```python
from automata.tm.dtm import DTM ## Importar librería "Automata" para máquinas de Turing.
```

```python
def DOm(): ## MT que genera la escala musical DO menor.
    EsDom = DTM(
    states={'q0', 'q1', 'q2', 'q3', 'q4', 'q5', 'q6', 'q7', 'q8'}, # Estados
    # Símbolos de entrada y símbolos de salida (notas musicales).
    input_symbols={'A ','A#Bb ','B ','C ','C#Db ','D ','D#Eb ','E ','F ','F#Gb ','G ','G#Ab '},
    tape_symbols={'A ','A#Bb ','B ','C ','C#Db ','D ','D#Eb ','E ','F ','F#Gb ','G ','G#Ab ',''}, #'' es blanco en la MT.
    transitions={
        'q0': {
            '': ('q1', 'C ', 'R')
        },
        'q1': {
            '': ('q2', 'D ', 'R')
        },
        'q2': {
            '': ('q3', 'D#Eb ', 'R')
        },
        'q3': {
            '': ('q4', 'F ', 'R')
        },
        'q4': {
            '': ('q5', 'G ', 'R')
        },
        'q5': {
            '': ('q6', 'G#Ab ', 'R')
        },
        'q6': {
            '': ('q7', 'A#Bb ', 'R')
        },
        'q7': {
            '': ('q8', 'C ', 'R')
        }  
    },
    initial_state='q0',
    blank_symbol='',
    final_states={'q8'}  
)
    return EsDom
```

## MENÚ - GENERADOR DE ESCALAS MENORES

```python
import IPython
def audio_escala(ubicacion): ## Reproducir escala automáticamente mediante display.Audio de la librería IPython.
    audio = IPython.display.display(IPython.display.Audio(filename=ubicacion, autoplay=True))
    return audio

def menu(): ## Estructura del menú con las opciones de escalas musicales.

    print ("Selecciona una opción")
    print ("\tC = DO")
    print ("\tD = RE")
    print ("\tE = MI")
    print ("\tF = FA")
    print ("\tG = SOL")
    print ("\tA = LA")
    print ("\tB = SI")
    print ("\t9 - salir")
 
Salir=True ## Mantener el ciclo while para mostrar el menú hasta que el usuario cierre el ciclo (Salir=False).
while Salir:
    # Mostramos el menu
    menu()
 
    # solicituamos una opción al usuario
    opcionMenu = input("\tEscribe la ESCALA MENOR que quieres generar >> ")
 
    if opcionMenu=="C":
        print ("")
        escala = str(DOm().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/c-minor.mp3')
        print ("")
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="D":
        print ("")
        escala = str(REm().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/d-minor.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="E":
        print ("")
        escala = str(MIm().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/e-minor.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="F":
        print ("")
        escala = str(FAm().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/f-minor.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="G":
        print ("")
        escala = str(SOLm().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/g-minor.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="A":
        print ("")
        escala = str(LAm().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/a-minor.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="B":
        print ("")
        escala = str(SIm().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/b-minor.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="9":
        Salir = False
    else:
        print ("")
        input("No has pulsado ninguna opción correcta...\npulsa una tecla para continuar")
```
