## MENÚ - GENERADOR DE ESCALAS MAYORES

```python
import IPython
def audio_escala(ubicacion):  ## Reproducir escala automáticamente mediante display.Audio de la librería IPython.
    audio = IPython.display.display(IPython.display.Audio(filename=ubicacion, autoplay=True))
    return audio

def menu(): ## Estructura del menú con las opciones de escalas musicales.

    print ("Selecciona una opción")
    print ("\tC = DO")
    print ("\tD = RE")
    print ("\tE = MI")
    print ("\tF = FA")
    print ("\tG = SOL")
    print ("\tA = LA")
    print ("\tB = SI")
    print ("\t9 - salir")
 
Salir=True ## Mantener el ciclo while para mostrar el menú hasta que el usuario cierre el ciclo (Salir=False).
while Salir:
    # Mostramos el menu
    menu()
 
    # solicituamos una opción al usuario
    opcionMenu = input("\tEscribe la ESCALA MAYOR que quieres generar >> ")
 
    if opcionMenu=="C":
        print ("")
        escala = str(DOM().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/c-major.mp3')
        print ("")
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="D":
        print ("")
        escala = str(REM().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/d-major.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="E":
        print ("")
        escala = str(MIM().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/e-major.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="F":
        print ("")
        escala = str(FAM().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/f-major.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="G":
        print ("")
        escala = str(SOLM().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/g-major.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="A":
        print ("")
        escala = str(LAM().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/a-major.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="B":
        print ("")
        escala = str(SIM().validate_input('')[1])
        print(escala[6:])
        audio_escala('audios/b-major.mp3')
        input("\npulsa una tecla para continuar")
    elif opcionMenu=="9":
        Salir = False
    else:
        print ("")
        input("No has pulsado ninguna opción correcta...\npulsa una tecla para continuar")
```
